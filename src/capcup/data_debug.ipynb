```python
%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np

import serial_data_formatter as sdf

cap_file = "data/translation_gt/251013_1409_randomTranslations_1000x_40mm.csv"
serial_data = sdf.SerialData(cap_file)
```

```python
time_periods = np.diff(serial_data.time)
fig, ax = plt.subplots()
ax.plot(serial_data.time[:-1], time_periods)
ax.set_xlabel("Time [s]")
ax.set_ylabel("Time between samples [s]")
ax.set_title("Time between received samples")
```

```python
fig, ax = plt.subplots()
ax.plot(serial_data.time, serial_data.cap_counts)
```
